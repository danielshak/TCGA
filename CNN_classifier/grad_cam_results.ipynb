```python
with open('Average_ggcam_train/avg_cam_maps_trn.pkl', 'rb') as f:
    train_maps = pickle.load(f)
```

```python
gene_names = pd.read_csv('train_genes.csv',header=None)
gene_names = gene_names.values.squeeze()
```

```python
top_genes = pd.DataFrame()
for i, img in enumerate(train_maps):
    indexes = np.argsort(np.reshape(img,(102*102))) #sorts from smallest to biggest, hence take last elements
    df = pd.DataFrame({cancer_names[i][0]: gene_names[np.flip(indexes[-10::])]})
    top_genes = pd.concat([top_genes,df],axis=1)
```
